## NFL Offense Data

1. Import the NFL week 3, 4, and 5 data and add a week column to dataset and set the value to 3, 4, or 5 for all rows from that week.

```python
import pandas as pd
```

```python
nfl3 = pd.read_csv('nfloffenseweek3.csv')
for row in nfl3:
    nfl3["week"] = 3

nfl4 = pd.read_csv('nfloffenseweek4.csv')
for row in nfl4:
    nfl4["week"] = 4

nfl5 = pd.read_csv('nfloffenseweek5.csv')
for row in nfl5:
    nfl5["week"] = 5
```

2. Combine the weekly dataframes by adding the dataframes together row-wise.

```python
all_weeks = pd.concat([nfl3, nfl4, nfl5])
all_weeks
```

3. Make up 1 interesting question about this dataset. Answer it.

What percent of passes were thrown and caught by the rank 1 team of each week?

```python
all_weeks["rank"] = all_weeks["rank"].astype(int)
rank1 = all_weeks[all_weeks["rank"] == 1]

total_attempted = rank1.passes_attempted.sum()
total_completed = rank1.passes_completed.sum()

percent = (((total_completed)*100) / total_attempted)
print(percent, "%")
```

## World Population Data

4. Join the height and weight data to the iq data. 
5. Join population density to quality of life data.
6. Join life expectancy to quality of life data.
7. Join iq, life expectancy, quality of life and population density into 1 dataframe.

```python
height_weight = pd.read_csv("height_weight_data.csv")
iq = pd.read_csv("iq.csv")
life_expect  = pd.read_csv("life_expectancy.csv")
population = pd.read_csv("population_density.csv")
quality_life = pd.read_csv("quality_of_life.csv")
```

```python
height_weight_iq = height_weight.merge(iq, on='country', how="right")
height_weight_iq
```

```python
population_quality = population.merge(quality_life, on="country", how="right")
population_quality
```

```python
all_country_data = height_weight_iq.merge(population_quality, on="country", how="left")
all_country_data
```

8. Make up 1 interesting question about one of these datasets and answer it.

Which country's with an iq of 90 or higher had the lowest popularity?

```python
high_iq = all_country_data[all_country_data.iq >= 90]

lowest_pop = high_iq.loc[high_iq.popularity.min()]
lowest_pop.country

```

9. Make up 1 interesting question about one of these datasets and answer it.

What is the average BMI of women who live in countries where females weigh more than males on average? What is the average bmi of men who live in those countries? vice versa?

```python
weights = all_country_data[all_country_data.female_weight > all_country_data.male_weight]
print("Female BMI in countries where women weigh more on average:", weights.female_bmi.mean())
print("Male BMI in countries where women weigh more on average:", weights.male_bmi.mean())

weights2 = all_country_data[all_country_data.female_weight < all_country_data.male_weight]
print("Female BMI in countries where men weigh more on average:", weights2.female_bmi.mean())
print("Male BMI in countries where men weigh more on average:", weights2.male_bmi.mean())
```

10. Make up 1 interesting question about one of these datasets and answer it.

For countries in the lower quarter of both health and safety, what is the difference between their highest iq and te highest iq of all countries?

```python
health_quarter = all_country_data.health.quantile(.25)
safety_quarter = all_country_data.safety.quantile(.25)

unhealthy_countries = all_country_data[(all_country_data.health < health_quarter) & (all_country_data.safety < safety_quarter)]

difference = all_country_data.iq.max() - unhealthy_countries.iq.max()
difference
```
